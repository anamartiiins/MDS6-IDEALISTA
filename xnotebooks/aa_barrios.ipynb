```python
import pandas as pd
from shapely import wkt
from shapely.geometry import Point
from shapely.geometry import Point
import geopandas as gpd
```

```python

# Cargar los datasets
sales_data_path = 'data\df_madrid_new_columns.csv'
polygons_data_path = 'data\Madrid_polygons.csv'

sales_data = pd.read_csv(sales_data_path)
polygons_data = pd.read_csv(polygons_data_path, delimiter=";")
```

```python
sales_data.head()
```

```python
polygons_data.head()
```

```python
# Convertir la columna WKT a geometrÃ­as utilizando geopandas
polygons_data['geometry'] = polygons_data['WKT'].apply(wkt.loads)
gdf_polygons = gpd.GeoDataFrame(polygons_data, geometry='geometry')

# Visualizamos los primeros registros de gdf_polygons para asegurarnos de que se ha convertido correctamente
gdf_polygons.head()
```

```python


# Convertir las coordenadas LATITUDE y LONGITUDE de sales_data en puntos geomÃ©tricos
sales_data['geometry'] = sales_data.apply(lambda row: Point(row['longitud'], row['latitud']), axis=1)

# Crear un GeoDataFrame para sales_data
gdf_sales = gpd.GeoDataFrame(sales_data, geometry='geometry')

# Visualizamos los primeros registros de gdf_sales para asegurarnos de que se ha convertido correctamente
gdf_sales.head()
```

```python
# Reiniciar los Ã­ndices de los GeoDataFrames
gdf_sales = gdf_sales.reset_index(drop=True)
gdf_polygons = gdf_polygons.reset_index(drop=True)
# Realizar la operaciÃ³n espacial para asociar cada punto con un polÃ­gono
# Usaremos 'predicate="within"' para encontrar los puntos que estÃ¡n dentro de los polÃ­gonos de los barrios
gdf_joined = gpd.sjoin(gdf_sales, gdf_polygons, how="left", predicate='within')

```
